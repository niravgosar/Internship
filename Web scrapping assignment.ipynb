### 2) Write s python program to display list of respected former presidents of India(i.e. Name , Term of office) from https://presidentofindia.nic.in/former-presidents.htm and make data frame

```python
page4=requests.get('https://presidentofindia.nic.in/former-presidents.htm')
soup5=BeautifulSoup(page4.content)
name=[]
years=[]
for i in soup5.find_all('div',class_="presidentListing"):
    name.append(i.text.split('\n')[1])
    years.append(i.text.split('\n')[2].split(':')[1])
ans2=pd.DataFrame({'Name':name, 'Term of Office':years})
```

### 3) a) Top 10 ODI teams in men’s cricket along with the records for matches, points and rating

```python
page4=requests.get('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
soup5=BeautifulSoup(page4.content)
teams=[]
matches=[]
points=[]
ratings=[]
records=[]
for i in soup5.find_all('span', class_="u-hide-phablet"):
    teams.append(i.text)
b=soup5.find('td', class_="rankings-block__banner--matches")
matches.append(b.text)

c=soup5.find('td', class_="rankings-block__banner--points")
points.append(c.text)
d=soup5.find('td', class_="rankings-block__banner--rating u-text-right")
ratings.append(d.text.split('\n')[1].lstrip())
for i in soup5.find_all('td', class_="table-body__cell u-center-text"):
    records.append(i.text)
for i in soup5.find_all('td', class_="table-body__cell u-text-right rating"):
    ratings.append(i.text)
for i in range(0, len(records), 1):
    if (i%2==0):
        matches.append(records[i])
    else:
        points.append(records[i])
ans3a=pd.DataFrame({'Teams':teams[0:10], 'Matches':matches[0:10], 'Points':points[0:10], 'Rating':ratings[0:10]})
ans3a.index=range(1,ans3a.shape[0]+1)
```

```python
ans3a
```

### 5) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world and make data frame-

```python
page=requests.get('https://www.cnbc.com/world/?region=world')
soup=BeautifulSoup(page.content)
news=[]
time=[]
link=[]
for i in soup.find_all('a',class_="LatestNews-headline"):
    news.append(i.text)
for i in soup.find_all('time', class_="LatestNews-timestamp"):
    time.append(i.text)
for i in soup.find_all('a',class_="LatestNews-headline"):
    link.append(i.get('href'))
ans5=pd.DataFrame({'Headline':news, 'Time':time, 'News link':link})
```

```python
ans5
```

### 7) Write a python program to scrape mentioned details from dineout.co.in and make data frame

```python
page=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
page
```

```python
soup=BeautifulSoup(page.content)
```

```python
names=[]
for i in soup.find_all('a', class_="restnt-name ellipsis"):
    names.append(i.text)
```

```python
cuisine=[]
for i in soup.find_all('span', class_="double-line-ellipsis"):
    cuisine.append(i.text.split('|')[1])
```

```python
loc=[]
for i in soup.find_all('div', class_="restnt-loc ellipsis"):
    loc.append(i.text)
```

```python
rat=[]
for i in soup.find_all('div', class_="restnt-rating rating-4"):
    rat.append(i.text)
```

```python
img=[]
for i in soup.find_all('img', class_="no-img"):
    img.append(i.get('data-src'))
```

```python
import pandas as pd
```

```python
ans7=pd.DataFrame({'Restaurant_name':names, 'Cuisine':cuisine, 'Location':loc, 'Ratings':rat, 'Image_URL':img})

ans7
```
