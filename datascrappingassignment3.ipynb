### In the above question, now scrape the following details of each product listed in first 3 pages of your search results and save it in a data frame and csv. In case if any product has less than 3 pages in search results then scrape all the products available under that product name. Details to be scraped are: "Brand Name", "Name of the Product", "Price", "Return/Exchange", "Expected Delivery", "Availability" and “Product URL”. In case, if any of the details are missing for any of the product then replace it by “-“.

```python
item_url=[]
brand=[]
name=[]
price=[]
ret=[]
delivery=[]
available=[]
```

```python
for page in range(0,3):
    url=driver.find_elements(By.XPATH,'//a[@class="a-link-normal s-underline-text s-underline-link-text s-link-style a-text-normal"]')
    for i in url:
        it=i.get_attribute('href')
        item_url.append(it)
    next_btn=driver.find_elements(By.XPATH,'//a[@class="s-pagination-item s-pagination-next s-pagination-button s-pagination-separator"]')
```

```python
len(item_url)
```

```python
for i in item_url:
    driver.get(i)
    time.sleep(1)
    try:
        brand_tag=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[46]/div/table/tbody/tr[1]/td[2]/span')
        brand.append(brand_tag.text)
    except NoSuchElementException :
        brand.append('-')
    try:
        name_tag=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[1]/div/h1/span')
        name.append(name_tag.text)
    except NoSuchElementException :
        name.append('-')
    try:
        price_tag=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[12]/div[3]/div[1]/span[3]/span[2]/span[2]')
        price.append(price_tag.text)
    except NoSuchElementException :
        price.append('-')
    try:
        return_tag=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[4]/div[24]/div[2]/div/div/div/div[2]/div/ol/li[3]/div/span/div[2]/span')
        ret.append(return_tag.text)
    except NoSuchElementException :
        ret.append('-')
    try:
        delivery_tag=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[3]/div/div[2]/div[9]/div[1]/div/div/div[1]/span/span')
        delivery.append(delivery_tag.text)
    except NoSuchElementException :
        delivery.append('-')
    try:
        avl_tag=driver.find_element(By.XPATH,'/html/body/div[2]/div[2]/div[5]/div[3]/div[1]/div[3]/div/div[1]/div/div/div/form/div/div/div/div/div[3]/div/div[4]/div/div[1]/span')
        available.append(avl_tag.text)
    except NoSuchElementException :
        available.append('-')
```

```python
ans2=pd.DataFrame({'Brand Name':brand, 'Product Name':name, 'Price':price, 'Return/Exchange':ret, 'Expected Delivery':delivery, 'Availability':available, 'Product URL':item_url})
```

### Write a python program to search for a smartphone(e.g.: Oneplus Nord, pixel 4A, etc.) on www.flipkart.com and scrape following details for all the search results displayed on 1st page. Details to be scraped: “Brand Name”, “Smartphone name”, “Colour”, “RAM”, “Storage(ROM)”, “Primary Camera”, “Secondary Camera”, “Display Size”, “Battery Capacity”, “Price”, “Product URL”. Incase if any of the details is missing then replace it by “- “. Save your results in a dataframe and CSV

```python
driver = webdriver.Chrome()
driver.get("http://www.flipkart.com/")
```

```python
a=input("Enter the smartphone brand")
```

```python
find=driver.find_element(By.XPATH, '/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/div/input')
find.send_keys(a)
search=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[1]/div[1]/div[2]/div[2]/form/div/button')
search.click()
```

```python
item_url=[]
brand=[]
name=[]
price=[]
colour=[]
ram=[]
rom=[]
cam=[]
dis=[]
battery=[]
```

```python
url=driver.find_elements(By.XPATH,'//a[@class="s1Q9rs"]')
for i in url:
    it=i.get_attribute('href')
    item_url.append(it)
```

```python
import regex as re
```

```python
for i in item_url:
    brand.append(a)
    driver.get(i)
    time.sleep(3)
    try:
        name_tag=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[4]/div/div[2]/div/div[1]/table/tbody/tr[3]/td[2]/ul/li')
        name.append(name_tag.text)
    except NoSuchElementException :
        name.append('-')
    try:
        color_tag=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[8]/div[4]/div/div[2]/div/div[1]/table/tbody/tr[4]/td[2]/ul/li')
        colour.append(color_tag.text)
    except NoSuchElementException :
        colour.append('-')
    try:
        r_tag=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[1]')
        pattern="[0-9]*"
        r=re.findall(pattern,r_tag.text)
        ram.append(r[0] +"GB")
    except NoSuchElementException :
        ram.append('-')
    try:
        r_tag=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[1]')
        pattern="[0-9]+"
        r=re.findall(pattern,r_tag.text)
        rom.append(r[1]+'GB')
    except NoSuchElementException :
        rom.append('-')
    try:
        ctag=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[3]')
        pattern="[0-9]*"
        c=re.findall(pattern,ctag.text)
        cam.append(c[0] +"MP")
    except NoSuchElementException :
        cam.append('-')
    try:
        dag=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[2]')
        pattern="[0-9]*"
        c=re.findall(pattern,dag.text)
        dis.append(c[0]+'.'+c[2]+'cm')
    except NoSuchElementException :
        dis.append('-')
    try:
        bat_tag=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[7]/div[1]/div/div[2]/ul/li[4]')
        pattern="[0-9]+"
        b=re.findall(pattern,bat_tag.text)
        battery.append(b[0] + 'mAh')
    except NoSuchElementException:
        battery.append('-')
    try:
        price_tag=driver.find_element(By.XPATH,'/html/body/div[1]/div/div[3]/div[1]/div[2]/div[2]/div/div[3]/div[1]/div/div[1]')
        price.append(price_tag.text)
    except NoSuchElementException:
        price.append('-')
```

```python
ans4=pd.DataFrame({'Brand Name':brand, 'Smartphone name': name, 'Colour':colour, 'RAM':ram, 'Storage(ROM)':rom, 'Primary Camera':cam,
'Display Size':dis, 'Battery Capacity':battery, 'Price':price, 'Product URL':item_url})
```

```python
ans4.to_csv('a4.csv')
```
